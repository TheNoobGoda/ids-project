# Business Understanding

  We are dealing with the telecommunications service provider industry, an industry that has traditionally provided telephone and similar services. In this industry there is a concept named ‘churn’ that refers to a phenomenon where subscribers of a telecommunications service provider discontinue their services or switch to another provider, usually measured as the rate at which customers cease their subscription. It would be beneficial for companies to know which customers will churn in order to prevent it from happening, as churn can cause a big financial impact.
  
  Therefore, **our business goal is to keep customers who will churn to continue using the services provided by this company.**
  
  In our case, **the data science goal is to predict if a customer will churn or not, based on certain attributes we know about them.**
  
  Achieving good predictions will allow the company to better **target their campaigns** to keep people using their services.
